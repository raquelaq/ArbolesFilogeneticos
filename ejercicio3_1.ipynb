```python
import random
```

En este ejercicio se ha desarrollado un script en Python para generar automáticamente un conjunto de secuencias de ADN aleatorias. Cada secuencia está compuesta por cinco nucleótidos elegidos al azar entre A, T, C y G, y se almacena en formato FASTA, que es el estándar para el intercambio de secuencias biológicas.

La generación automática de secuencias permite disponer de datos sintéticos controlados, útiles para ilustrar el funcionamiento de los métodos filogenéticos sin depender de datos reales ni introducir sesgos biológicos adicionales.

```python
nucleotides = ["A", "T", "C", "G"]

num_sequences = 5
sequence_length = 5

output_file = "random_sequences.fasta"

with open(output_file, "w") as f:
    for i in range(num_sequences):
        species_name = f"Especie_{chr(65 + i)}"  # A, B, C, D, E
        sequence = "".join(random.choice(nucleotides) for _ in range(sequence_length))
        
        f.write(f">{species_name}\n")
        f.write(sequence + "\n")

print(f"Archivo FASTA generado: {output_file}")
```

Dado que las secuencias se generan de forma aleatoria, no se espera una señal filogenética clara, lo que permite observar el comportamiento de los algoritmos en ausencia de relaciones evolutivas reales.
